```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.feature_names
```

```python
X = iris.data[:, 2:]
X.shape
```

```python
Sigma = np.cov(X, rowvar=0)
Sigma
```

```python
_, S = np.linalg.eig(Sigma)
S
```

```python
y = np.dot(S.T, X.T).T
y.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
plt.scatter(y[:, 0], y[:, 1])
plt.show()
```
