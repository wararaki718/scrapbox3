```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler, StandardScaler
```

```python
np.random.seed(seed=42)
```

```python
X = np.random.multivariate_normal([5, 5], [[5, 0], [0, 2]], 50)
```

```python
X_mas = MaxAbsScaler().fit_transform(X)
```

```python
X_mms = MinMaxScaler().fit_transform(X)
```

```python
X_ss = StandardScaler().fit_transform(X)
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X[:, 0], X[:, 1], c='red', marker='x')
ax.scatter(X_mas[:, 0], X_mas[:, 1], c='blue', marker='x')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X[:, 0], X[:, 1], c='red', marker='x')
ax.scatter(X_mms[:, 0], X_mms[:, 1], c='blue', marker='x')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X[:, 0], X[:, 1], c='red', marker='x')
ax.scatter(X_ss[:, 0], X_ss[:, 1], c='blue', marker='x')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X_mas[:, 0], X_mas[:, 1], c='red', marker='x')
ax.scatter(X_mms[:, 0], X_mms[:, 1], c='blue', marker='x')
ax.scatter(X_ss[:, 0], X_ss[:, 1], c='green', marker='x')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(X_mas[:, 0], X_mas[:, 1], c='red', marker='x')
ax.scatter(X_mms[:, 0], X_mms[:, 1], c='blue', marker='x')
plt.show()
```
